<h3> EDA

<h4> what is the overall sales trends?
    

```python
 df['order_date'].min()
```

```python
df['order_date'].max()
```

```python
# create a new column with formatted date string
df['month_year'] = df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
monthly_trends = df.groupby('month_year').sum()['sales'].reset_index()
monthly_trends
```

```python
sns.set_style('ticks')
```

```python
plt.figure(figsize=(15,6))
plt.plot(monthly_trends['month_year'],monthly_trends['sales'])
plt.xticks(rotation='vertical',size=8)
plt.show()
```

- <h4> what are the top 10 product by sales?

```python
prod_sales = pd.DataFrame(df.groupby('product_name').sum()['sales'])
prod_sales
```

```python
top_10_prod = prod_sales.sort_values(by='sales', ascending=False).head(10)
top_10_prod
```

- <h4> what are most selling prodcuts ?

```python
most_selling=pd.DataFrame(df.groupby('product_name').sum()['quantity'])
```

```python
most_selling.sort_values('quantity',ascending=False).head(10)
```

- <h4> what are the most prefered ship mode >

```python
sns.countplot(df['ship_mode'])
```

- <h4> what are the most profitable category and sub_category ?

```python
most_proft_cat=pd.DataFrame(df.groupby(['category','sub_category']).sum()['profit'])
```

```python
most_proft_cat.sort_values(['category','profit'],ascending=False)
```
